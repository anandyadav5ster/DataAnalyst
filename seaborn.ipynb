```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

### seaborn is used to create statistical graphics in Python.
like groups, class
Relational plots
Catagorial plots
Distribution plots


```python
var = [1,2,3,4,5]
var1= [6,7,8,9,10]
plt.plot(var,var1)
plt.show()
```

```python
sns.lineplot(x=var, y=var1)
plt.show()
```

```python
var = [1,2,3,4,5]
var1= [6,7,8,9,10]
df = pd.DataFrame({'var':var,'var1':var1})
df
```

```python
sns.lineplot(data=df,x='var', y ='var1')
plt.show()
```

```python
url = 'https://raw.githubusercontent.com/mwaskom/seaborn-data/master/penguins.csv'
# url = 'https://github.com/mwaskom/seaborn-data/blob/master/penguins.csv'
df = sns.load_dataset('penguins')
```

```python
df.head()
```

```python
sns.lineplot(x= 'bill_length_mm',y='flipper_length_mm',data=df)
plt.show()
```

```python
# create categorial plot with sex as hue
sns.lineplot(x= 'bill_length_mm',y='flipper_length_mm',data=df,hue='sex')
plt.show()
```

```python
# customize the plot with markers and style
sns.lineplot(x= 'bill_length_mm',y='flipper_length_mm',data=df,hue='sex', palette='rocket_r')
plt.show()
```

```python
# set marker and style
sns.lineplot(x= 'bill_length_mm',y='flipper_length_mm',data=df,hue='sex', markers=['o','>'],style='sex')
plt.show()
```

```python
df1= sns.load_dataset('penguins').head(20)
df1
```

```python
sns.lineplot(x= 'bill_length_mm',y='flipper_length_mm',data=df1,hue='sex', markers=['o','>'],style='sex')
plt.show()
```

```python
# add grid to the plot
sns.lineplot(x= 'bill_length_mm',y='flipper_length_mm',data=df1,hue='sex', markers=['o','>'],style='sex')
plt.title('Penguin Bill Length vs Flipper Length')
plt.grid()
plt.show()
```
