**Adding values on top of graph**

```python

x = np.array(['A','B','C','D'])
y= np.array([3,8,1,10])
plt.xlabel('Catagory')
plt.ylabel('Values')
plt.title('Horizontal Bar Chart with Values')
plt.barh(x,y, color='g',height=0.5)
# Adding values on top of graph
for i,v in enumerate(y):
    # plt.text(x_position, y_position, text)
    # We add 0.1 to v so the text doesn't touch the bar
    plt.text(v+0.1,i,str(v), va='center',fontweight='normal')
# Adjust the x-limit so the text doesn't get cut off
plt.xlim(0,max(y)+1)    
plt.show()
```

**Scatter Plot**

```python
x = np.array(['A','B','C','D'])
y= np.array([3,8,1,10])
plt.title('Scatter Plot')
for i,v in enumerate(y):
    plt.text(x[i],v+0.5,str(v),va='top',fontweight='bold')
c = ['red','magenta','green','orange']    
plt.scatter(x,y,color=c,s =100)
plt.show()
```

```python
# Pie chart
subject = ['English','Hindi','Chemistry','Maths']
sizes = [15,30,45,10]
# Displays the percentage on the slices autopct='%1.1f%%'
plt.pie(sizes,labels=subject,
        autopct='%1.1f%%')
plt.axis('equal')
plt.show()
```

```python
# Pie chart
subject = ['English','Hindi','Chemistry','Maths']
sizes = [15,30,45,10]
c = ['r','g','b','y']
# Displays the percentage on the slices autopct='%1.1f%%'
plt.pie(sizes,labels=subject,
        autopct='%1.1f%%',colors=c)
plt.axis('equal')
plt.show()
```

```python
# Pie chart
subject = ['English','Hindi','Chemistry','Maths']
sizes = [15,30,45,10]
c = ['r','g','b','y']
# Displays the percentage on the slices autopct='%1.1f%%'
plt.pie(sizes,labels=subject,
        autopct='%1.1f%%',colors=c)
plt.axis('equal')
plt.legend(title='Subjects',loc='upper right', bbox_to_anchor=(1,0,0.4, 1))
plt.show()
```
