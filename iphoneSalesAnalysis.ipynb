```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.graph_objects as go
import plotly.express as px
```

```python
df = pd.read_csv('apple_products.csv')
df.head()
```

Fing top 10 iphone data in india

```python
hightest_rated = df.sort_values(by='Star Rating', ascending=False).head(10)
hightest_rated['Product Name']
```

lets have a look for the highest rated iphone on flipkart

```python
iphones = hightest_rated["Product Name"].value_counts()
iphones = iphones.index
counts = hightest_rated["Number Of Ratings"]
```

```python
figure = px.bar(x=iphones,y=counts,title="Top 10 Highest Rated iPhones",labels={"x":"iPhone Models","y":"Number of Ratings"})
figure.show()
```

```python
# Review plots and further analysis can be added here
iphones = hightest_rated['Product Name'].value_counts()
labels = iphones.index
counts = hightest_rated['Number Of Reviews']
figure = px.bar(hightest_rated,x=labels,y=counts,title="Top 10 Highest Rated iPhones",labels={"x":"iPhone Models","y":"Number of Reviews"})
figure.show()
```

```python
# plot and further analysis can be added here for rating and price
figure = px.scatter(data_frame=df,x='Number Of Ratings',y='Sale Price', size = 'Discount Percentage', trendline="ols",color='Star Rating', title='iPhone Sales Price vs Number of Ratings')
figure.show()
```

```python
# vistualizations and further analysis can be added here discount and rating
figure = px.scatter(data_frame=df,x='Number Of Ratings',y='Discount Percentage', size = 'Sale Price', trendline="ols",color='Star Rating', title='iPhone Discount Percentage vs Number of Ratings')
figure.show()
```
